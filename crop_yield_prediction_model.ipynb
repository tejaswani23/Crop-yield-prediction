```python
import pandas as pd
```

```python
df=pd.read_csv("crop_yield_df.csv")
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['Area'].unique()
```

```python
df['Item'].unique()
```

```python
df['Year'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
area=l.fit_transform(df['Area'])
res = {}
for cl in l.classes_:
        res.update({cl:l.transform([cl])[0]})
res
```

```python
df['area']=area
df=df.drop('Area',axis=1)
df
```

```python
df.dtypes
```

```python
item=l.fit_transform(df['Item'])
item
```

```python
df['item']=item
df=df.drop('Item',axis=1)
df
```

```python
X=df.drop('hg/ha_yield',axis=1)
X
```

```python
Y=df['hg/ha_yield']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(X_train,y_train)
yield1=model1.predict(X_test)
yield1
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
model2=RandomForestRegressor()
model2.fit(X_train,y_train)
yield2=model2.predict(X_test)
yield2
```

```python
r2_score(yield1,y_test)#LR
```

```python
r2_score(yield2,y_test)#RFR
```
